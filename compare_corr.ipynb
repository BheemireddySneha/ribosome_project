```python
fin,fin1,fin2,fin3,fin4,fin5,fin6 = [],[],[],[],[],[],[]
#f = open("demo_4j.txt", "a")
df_pdb = pd.read_csv('atm_4j.csv',header=None,names = ['atm_num','atm_type','res_type','chain','res_num'],sep=',')

for i in range(0,len(df_cross_4j[0])):
    for j in range(0,len(df_cross_4j[0])):
        if df_cross_4j[i][j]>0.7:
            res1 = df_pdb['chain'][i]
            res2 = df_pdb['chain'][j]
            #print("***")
            if res1 != 'A' :
                if res2 != 'A':
                    if res1 != res2:
                        fin.append(df_cross_4j[i][j])
                        fin1.append(i)
                        fin2.append(j)
                        fin3.append(df_pdb['chain'][i])
                        fin4.append(df_pdb['res_num'][i])
                        fin5.append(df_pdb['chain'][j])
                        fin6.append(df_pdb['res_num'][j])
                
#f.close()
df_fin = pd.DataFrame()
df_fin['val'] = fin
df_fin['i'] =fin1
df_fin['j'] =fin2
df_fin['ch1'] = fin3
df_fin['ch2'] = fin5
df_fin['res_num1'] = fin4
df_fin['res_num2'] = fin6
df_fin.to_csv('4j_fin.csv')
#df_fin = df_fin[df_fin['ch2'] != 'A']
```

```python
pdb = parsePDB("4ji1_f.pdb")
#,chain='D')
#ca = pdb.select("name CA")
pdb = pdb.select("name CA or name P")
dist = pd.DataFrame(buildDistMatrix(pdb,format='mat'))
distance = []
for dat in range(0,len(df_fin['val'])):
    distance.append(dist[df_fin['i'][dat]][df_fin['j'][dat]])
df_fin['dist']=distance

df_fin.to_csv('4j_fin.csv')
```
